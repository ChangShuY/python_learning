```python
%matplotlib inline
from paddle.utils.plot import Ploter # 加载绘图函数
curve=Ploter('Line1','Line2','L3') # 定义一个图板
for i in range(10):
    curve.append('Line1',i,i*i) # 增加曲线上的点。参数：曲线名(必须定义图板时出现过)；x轴；y轴
    curve.append('Line2',i,10*i) # 增加曲线上的点。参数：曲线名(必须定义图板时出现过)；x轴；y轴
    curve.append('L3',2*i,i*i*i) # 增加曲线上的点。参数：曲线名(必须定义图板时出现过)；x轴；y轴
curve.plot() #绘图
```
